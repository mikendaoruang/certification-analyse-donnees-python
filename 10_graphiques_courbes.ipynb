## Comparaison de graphiques

```python
fig = plt.figure(figsize=(12,12))

# On répète la boucle 5 fois (i = 0, 1, 2, 3, 4) → une fois par année
for i in range(5):
    # On crée un graphique dans la figure : 5 lignes, 1 colonne, à la position i+1
    ax = fig.add_subplot(5,1,i+1)
    # On calcule le début de la tranche : 0, 12, 24, 36, 48
    start_index = i*12
    # On calcule la fin de la tranche : 12, 24, 36, 48, 60
    end_index = (i+1)*12
    # On extrait 12 lignes (= 1 an de données) depuis le tableau unrate
    subset = unrate[start_index:end_index]
    # On trace la courbe : DATE en axe X, taux de chômage en axe Y
    ax.plot(subset['DATE'], subset['UNRATE'])

plt.show()
```

## Ajouter plus de courbes

```python
# Régler les dimensions 
fig = plt.figure(figsize=(10,6))

# Création d'une liste avec les différentes couleurs
colors = ['red', 'blue', 'green', 'orange', 'black']

# Création d'une boucle for 
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]

    plt.plot(subset['MONTH'],subset['UNRATE'], c=colors[i])

plt.show()
```

### Exercice

```python
fig = plt.figure(figsize=(10,6))

colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    # Définir un label de 1948, ainsi de suite avec i par rapport à notre boucle
    label = str(1948 + i)
    plt.plot(subset['MONTH'],subset['UNRATE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))

colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    # Définir un label de 1948, ainsi de suite avec i par rapport à notre boucle
    label = str(1948 + i)
    plt.plot(subset['MONTH'],subset['UNRATE'], c=colors[i], label=label)
plt.legend(loc='upper left')

# Ajout d'un titre au graphique 
plt.title("Evolution du taux de chômage de 1948 à 1952")

# Ajout des intitulé sur les axes
plt.xlabel('Mois (de 1 à 12')
plt.ylabel('Taux de chômage (en %)')

plt.show()
```
