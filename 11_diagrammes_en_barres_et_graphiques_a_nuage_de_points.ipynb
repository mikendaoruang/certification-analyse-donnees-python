## Introduction au dataset

```python
import pandas as pd
reviews = pd.read_csv("fandango_scores.csv")

# Assigner les colonnes du dataframe à la variable cols
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

# Assigner les colonnes à la variable norm_reviews
norm_reviews = reviews[cols]

# Afficher les 3 premiéres lignes avec les colonnes demandées
print(norm_reviews[:3])
```

```python
# Création du sous graphique
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 1
tick_positions = range(1,6)

# Création du diagramme
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols,rotation=90)

# Titres sur les graphiques 
ax.set_xlabel('Sources de notation')
ax.set_ylabel('Note moyenne')
ax.set_title('Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)')

plt.show()
```

### Exercice

```python
import matplotlib.pyplot as plt 

# Importer la fonction arange dans Numpy
from numpy import arange

nom_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_widths = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 1
tick_positions = range(1,6)
fig, ax = plt.subplots()

ax.barh(bar_positions, bar_widths, 0.5)

ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)

# Titres sur les graphiques 
ax.set_xlabel('Sources de notation')
ax.set_ylabel('Note moyenne')
ax.set_title('Moyenne des notes utilisateurs pour le film Avengers: Age of Ultron (2015)')

plt.show()
```

## Graphique à nuage de points

```python
# Axes.scatter(x, y) = permet de générer un graphique à nuage de points
```

```python
# Création d'un sous-graphique
fig, ax = plt.subplots()

# Générer un graphique à nuage de points
ax.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])

# Définir les titres du graphique 
ax.set_xlabel('Fangando')
ax.set_ylabel('Rotten Tomatoes')

# Afficher le graphique
plt.show()
```

## Intervertir les axes

```python
# Création de sous-graphique 
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

# Création du graphique à nuage de points pour l'ax1
ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])
ax1.set_xlabel('Fandango')
ax1.set_ylabel('Rotten Tomatoes')

# Création du graphique à nuage de points pour l'ax2
ax2.scatter(norm_reviews['RT_user_norm'], norm_reviews['Fandango_Ratingvalue'])
ax2.set_xlabel('Rotten Tomatoes')
ax2.set_ylabel('Fandango')

# Affichage des graphiques
plt.show()
```

## Comparaison de corrélations

```python
# Axes.set_xlim()

# Axes.set_ylim()

# Ces fonctions vont nous permettre de définir les limites sur chaque axes
```

```python
ax.set_xlim(0,5)
ax.set_ylim(0,5)
```

```python
import matplotlib.pyplot as plt

# Création du sous graphique 1
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

# Création du graphique à nuage de point de l'ax1
ax1.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['RT_user_norm'])

# Définir les titres du graphique à nuage de point
ax1.set_xlabel('Fandango')
ax1.set_ylabel('Rotten Tomatoes')

# Définir les limites des axes x et y en les fixant de 0 à 5 
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)

# Adapter pour l'ax2
ax2.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['Metacritic_user_nom'])

ax2.set_xlabel('Fandango')
ax2.set_ylabel('Metacritic')

ax2.set_xlim(0,5)
ax2.set_ylim(0,5)

# Adapter pour l'ax3
ax3.scatter(norm_reviews['Fandango_Ratingvalue'], norm_reviews['IMDB_norm'])

ax3.set_xlabel('Fandango')
ax3.set_ylabel('IMDB')

ax3.set_xlim(0,5)
ax3.set_ylim(0,5)

# Affichage du graphique 
plt.show()
```
