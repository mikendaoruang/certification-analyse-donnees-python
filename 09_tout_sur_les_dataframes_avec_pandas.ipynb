## Utiliser des index personnalisés

```python
# DataFrame.set_index(Series, inplace=True, drop=False)
# inplace=True : permet de remplacer le Series indiqué
# drop=False : la colonne spécifié en paramétre sera conservé 
```

```python
fandango_films = fandango.set_index("FILM", drop=False)
print(fandango_films.index)
```

## Méthode Apply() sur les colonnes d'un DataFrame

```python
# apply()
```

```python
# dataframe.apply(lambda x: x**2) : Nouveau data frame élevé à la puissance deux chaque élément des colonnes
```

```python
import numpy as np
```

```python
# retourne les types de données en tant que Objet Series
types = fandango_films.dtypes
types
```

```python
# filtrer les nombres à virgules float64 et retourner le nom des colonnes
float_columns = types[types.values == "float64"].index
float_columns
```

```python
# Utilisation des crochets pour filtrer les colonnes de la condition float_columns
float_df = fandango_films[float_columns]
float_df
# On a un DataFrame avec toutes les lignes, on aura seulement les colonnes avec des types de données float64
```

```python
# Calculer l'écart-type de chaque colonne du DataFrame float_df
deviation = float_df.apply(lambda x: np.std(x))
```

```python
print(deviation)
```

## Méthode Apply() sur les lignes d'un DataFrame

```python
# apply(function, axis=1)
```

```python
# Obtenir un DataFrame avec toutes les lignes et seulement 2 colonnes
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_user
```

```python
# Obtenir un objet series qui permet de savoir l'écart-type de chaque ligne
rt_mt_user.apply(lambda x:np.std(x), axis=1)
```
