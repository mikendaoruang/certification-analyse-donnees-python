## Boucles FOR

```python
cities = ['Paris', 'Madrid', 'Rome']
```

```python
for city in cities: 
    print(city)
```

```python
fruits = ["Citron", "Banane", "Pomme", "Banane", "Pomme", "Banane", "Pomme"]
```

```python
fruits_counts = {}
```

```python
for item in fruits: 
    if item in fruits_counts: 
        fruits_counts[item] += 1 
    else:
        fruits_counts[item] = 1
```

```python
print(fruits_counts)
```

## Fonctions

```python
def ton_age(age): 
    if age >= 60:
        return("âgé")
    else:
        return("jeune")
```

```python
ton_age(90)
```

```python
ton_age(20)
```

```python
def listing(dict): 
    liste = []
    for d in dict: 
        if dict[d] > 10:
            liste.append(d)
        return liste
```

```python
notes = {
    "Tom" : 12,
    "Julie" : 19,
    "Nina" : 9,
    "Bob" : 5
}
```

```python
listing(notes)
```
