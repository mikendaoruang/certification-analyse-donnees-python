## Cas Pratique

```python
# Utiliser l'attribut columns pour retourner le nom des colonnes du dataframe food_info et convertir en liste utilisant la méthode tolist()
col_names = food_info.columns.tolist()

# Créer une nouvelle liste qu'on nommera gram_columns, contenant seulement les colonnes se terminant par "(g)"
gram_columns = []

# Boucle For
for c in col_names: 
    if c.endswith("(g)"):
        gram_columns.append(c)

# Création d'un nouveau dataframe food_info
gram_df = food_info[gram_columns]

# Afficher les 3 premères valeurs de ce dataframe gram_df
gram_df.head(3)
```
