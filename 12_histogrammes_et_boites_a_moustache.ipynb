# Histogrammes et boîtes à moustache

## Introduction

```python
import pandas as pd

reviews = pd.read_csv('fandango_scores.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

# Affichage des 5 premières lignes de cols dans le dataframe reviews
norm_reviews = reviews[cols]
print(norm_reviews[:5])
```

## Distribution de fréquences

```python
# pour retourner une distribution de fréquences = Series.value_counts()
```

```python
# Obtenir les valeurs les plus communes dans une colonne donnée 
freq_counts = norm_reviews['Fandango_Ratingvalue'].value_counts()
freq_counts 
```

```python
# Trier par index = Series.sort_index()
```

```python
# Trier par ordre croissant en fonction des notes 
sorted_freq_counts = freq_counts.sort_index()
sorted_freq_counts
```

## Exercice

```python
# Afficher les valeurs les plus communes de la colonne 'IMDB_norm'
imdb_distribution = norm_reviews['IMDB_norm'].value_counts()

# Afficher le résultat
imdb_distribution
```

```python
# Trier l'index notes par ordre croissant
sorted_imdb_distribution = imdb_distribution.sort_index()

# Afficher le résultat 
sorted_imdb_distribution
```

## Les histogrammes avec Matplotlib

```python
# Générer un histogramme = Axes.hist(values, range(a,b))
```

```python
# ax.hist(dt[col],range=[0,5])
```

```python
import matplotlib.pyplot as plt

# Création d'un histogramme
fig, ax = plt.subplots()
ax.hist(norm_reviews['Fandango_Ratingvalue'],range=[0,5])

# Afficher l'histogramme
plt.show()
```

## Comparaison d'histogrammes

```python
# ax.hist(Series, bins=20, range=(a,b)) ou ax.hist(Series, 20, range=(...))
```

```python
# Axes.set_ylim(0,25)
```

```python
# Création de sous graphique et définir les dimensions
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

# Histogramme des valeurs de la colonne 'Fandango_Ratingvalue' en utilisant 20 classes (bins), intervalle 0 à 5
ax1.hist(norm_reviews['Fandango_Ratingvalue'], bins=20, range=(0,5))

# Définir les titres
ax1.set_title('Distribution des notes de fandango')

# Définir l'intervalle sur l'axe y de 0 à 50
ax1.set_ylim(0,50)

# Reproduire pour l'ax2
ax2.hist(norm_reviews['RT_user_norm'], bins=20, range=(0,5))
ax2.set_title('Distribution des notes de Rotten Tomatoes')
ax2.set_ylim(0,50)

# Reproduire pour l'ax3
ax3.hist(norm_reviews['Metacritic_user_nom'], bins=20, range=(0,5))
ax3.set_title('Distribution des notes de Metacritic')
ax3.set_ylim(0,50)

# Reproduire pour l'ax3
ax4.hist(norm_reviews['IMDB_norm'], bins=20, range=(0,5))
ax4.set_title('Distribution des notes de IMBD')
ax4.set_ylim(0,50)

# Affichage des graphiques 
plt.show()
```

## Boîte à moustaches

```python
# axes.boxplot()
```

```python
fig, ax = plt.subplots()
ax.boxplot(norm_reviews['RT_user_norm'])
ax.set_xticklabels(['Rotten Tomatoes'])
ax.set_ylim(0,5)
plt.show()
```

## Plusieurs diagrammes à boîtes

```python
# sélection des colonnes dans le dataframe 
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

# Création du diagramme à boîte
fig, ax = plt.subplots()
ax.boxplot(norm_reviews[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0,5)

# Affichage du diagramme
plt.show()
```
