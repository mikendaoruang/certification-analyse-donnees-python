## Problème avec les valeurs manquantes

```python
age_is_null = pd.isnull(titanic_survival["age"])

# Afficher tous les âges qui sont présent
good_ages = titanic_survival["age"][age_is_null == False]

# Calcul de la moyenne d'âge
mean_age = sum(good_ages) / len(good_ages)
print(mean_age)
```

## Calculer des statistiques de prix

```python
fares_by_class = {}

passenger_classes = [1,2,3]

for this_class in passenger_classes:
    pclass_rows = titanic_survival[titanic_survival["pclass"] == this_class]
    pclass_fares = pclass_rows["fare"]
    fare_for_class = pclass_fares.mean()
    fares_by_class[this_class] = fare_for_class

print(fares_by_class)
    
```

## Appliquer des fonctions sur un DataFrame

```python
# DataFrame.apply()
```

```python
# Créer une fonction qui retourne le 100e élément

def row_100(column):
    # extraire le 100e élement d'une colonne
    item = column.iloc[99]
    return item
```

```python
# retourne la 100eme élément (ligne) de chaque colonne
row_100_var = titanic_survival.apply(row_100)
row_100_var
```

## Appliquer des fonctions sur un DataFrame

```python
# Définir un paramétres
def null_count(column):
    # Trouver les valeurs manquante avec la libraire pd et la méthode null
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)
```

```python
# Permettre de donner le nombre de valeur manquantes pour chaque colonne
column_null_count = titanic_survival.apply(null_count)
print(column_null_count)
```

## Appliquer une fonction à une ligne

```python
# DataFrame.apply(function, axis=1)
# axis=1 : Sur chaque ligne de notre DataFrame 
```

```python
# Création d'une fonction en définissant le paramétres lignes
def is_minor(row):
    if row["age"] < 18: 
        return True
    else:
        return False
```

```python
# Appliquer la fonction 
minors = titanic_survival.apply(is_minor, axis=1)
minors
```

```python
# Création d'une fonction
def which_class(row):
    # On va chercher les éléments de la colonne pclass
    pclass = row["pclass"]

# On vérifie si la valeur est manquante dans la colonne pclass avec la libraire pd.isnull ensuite on définie des conditions
    if pd.isnull(pclass): 
        return "Unknown"
    elif pclass == 1: 
        return "First Class"
    elif pclass == 2: 
        return "Second Class"
    else:
        return "Third Class"
```

```python
classes = titanic_survival.apply(which_class, axis=1)
classes
```

### Exercice

```python
def minor_age(row): 
    age = row["age"]

    if pd.isnull(age): 
        return "unknown"
    elif age < 18:
        return "minor"
    else:
        return "adult"
```

```python
age_def = titanic_survival.apply(minor_age, axis=1)
age_def
```

## Pratique: Calculer le pourcentage de survie par groupe d'âge

```python
# Ajouter une nouvelle colonne dans le dataframe
titanic_survival["age_def"] = age_def

#Créer un pivot table
age_group_survival = titanic_survival.pivot_table(index="age_def", values="survived")

print(age_group_survival)
```
